```python
#@title
import svgwrite as  svg,\
                    math,\
                    numpy,\
                    datetime,\
                    pandas
from dateutil.relativedelta import relativedelta

from svgwrite.shapes    import Rect,Line,Circle
from svgwrite.text      import Text
from svgwrite.path      import Path
from svgwrite.container import Group
from google.colab import files

class Plot:
    '''
    Description
    -----------
    This radial plot system enables plotting graphs on circle

    '''
    def __init__(self,
          #labels
          label_surface_size      : float  =.04 ,
          label_surface_position  : float = 270,
          #global style
          image_background_style  : dict  = {},
          image_height            : int   = 1000,
          image_width             : int   = 1100,
          image_size_unit         : str   = 'px',
          #circles size params
          circles_center          : tuple = (500,550),
          circles_max_radius_size : float = 420.0,
          circles_rings_positions : list  = [(.18,.20),
                                             (.20,.30),
                                             (.30,.40),
                                             (.40,.60),
                                             (.60,10)],
          circles_style           : dict  ={}


          )->None :
          self.label_surface_size       = label_surface_size
          self.label_surface_position   = label_surface_position
          self.image_background_style   = image_background_style
          self.image_height             = image_height
          self.image_width              = image_width
          self.image_size_unit          = image_size_unit
          self.circles_center           = circles_center
          self.circles_max_radius_size  = circles_max_radius_size
          self.circles_rings_positions  = circles_rings_positions
          self.circles_style           = circles_style


          self.circles_rings_radius_list= [*map(lambda x : (self.circles_max_radius_size*x[0],self.circles_max_radius_size*x[1]),self.circles_rings_positions)]

          #initializing my circles
          self.image                    = svg.Drawing(size = (f'{self.image_width}{self.image_size_unit}',f'{self.image_height}{self.image_size_unit}'))
          self.image.add(\
          Rect(\
          size=('100%', '100%'),\
          **self.image_background_style
            )
          )

          #circles utils
          self.translate              = lambda P,C=self.circles_center : (P[0]+C[0],P[1]+C[1])
          self.get_circle_coords      = lambda O,R : self.translate((R*math.cos(O),R*math.sin(O)))
          self.FromHalfPi             = lambda tetas :[O for O in tetas if O >= 270] + [O for O in tetas if O < 270]
          self.compute_angles         = lambda n,dn: list(O for O in numpy.linspace(0, 360, num=n+dn*2))

          #shards' draw utils
          self.drawboundaryline       = lambda  shard,linesstyle,key : Line(shard['coords'][key]['down'],shard['coords'][key]['up'],**linesstyle)
          self.shardlabelposition     = lambda  shard,key,label_left : (shard['coords'][key]['up'],(shard['coords'][key]['up'][0]-label_left,shard['coords'][key]['up'][1]))[ 90<shard['teta']<270]
          self.get_label              = lambda  shard,labelstyle,label_left,key='start' : Text(shard['id'][2],insert=self.shardlabelposition(shard,key,label_left),**labelstyle)
          self.euclid_dist            = lambda  p0,p1 :((p0[0]-p1[0])**2+(p0[1]-p1[1])**2)**0.5

          #draw paths
          self.pie                    = lambda pts,r0,r1 : f"M {' '.join(str(x)for x in pts[0])} L {' '.join(str(x) for x in pts[1])} A {r1},{r1},0,0,1, {' '.join(str(x) for x in pts[2])} L {' '.join(str(x)for x in pts[3])} A {r0},{r0},0,0,0, {' '.join(str(x) for x in pts[0])} "
          self.RectOnArc              = lambda pts,r0 : f"M {' L '.join(' '.join(str(x)for x in pi) for pi in pts)} A {r0},{r0},0,0,0, {' '.join(str(x) for x in pts[0])}"
          self.CustomRect             = lambda pts : f"M {' L '.join(' '.join(str(x)for x in pi) for pi in pts)} Z"

          #adding circles
          circles                     = [Circle(center = self.circles_center, r = self.circles_rings_radius_list[0][0],**self.circles_style)] +\
                                        [Circle(center = self.circles_center, r = r1,**self.circles_style) for (r0,r1) in self.circles_rings_radius_list]
          for c in circles            : self.image.add(c)
          self.n_circles              = len(circles)
    def human_format(self,num):
        magnitude = 0
        while abs(num) >= 1000:
            magnitude += 1
            num /= 1000.0
        return '%.0f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])
    def get_drawable_angles(self,n):
              dn                      = int(self.label_surface_size*n)
              tetas                   = self.compute_angles(n,dn)
              part1,part2             = [O for O in tetas if O <=self.label_surface_position],[O for O in tetas if O >self.label_surface_position]
              part1_cut               = part1[0:n-dn]
              part2_cut               = part2[dn:n]
              tetas                   = self.FromHalfPi(part1_cut + part2_cut)
              return tetas
    def curve(self,
                    pts,
                    boundaries,
                    r0,
                    t=1/3):
              point_to_str = lambda p,sep=' ' : sep.join(str(x) for x in p)
              (b00,b10,b11,b01) = boundaries
              def ctr(p,t,sep=' ',raw=False):
                  (x0,y0),(x,y),(x1,y1)= p
                  dx,dy = x0-x1,y0-y1
                  c0    = (x-dx*t, y-dy*t)
                  c1    = (x+dx*t, y+dy*t)
                  return (f" {point_to_str(c0,sep)} {point_to_str(c1,sep)} ",(c0,c1))[raw]
              d = f"M {point_to_str(b00)} "
              d+= f'L {point_to_str(b10)} '
              _,(x0,y0) = ctr(pts[0],t,sep=',',raw=True)
              d+= f"Q {point_to_str((x0,y0),sep=',')}, {point_to_str(pts[0][0],sep=',')} "
              for p in pts :
                d  += f"C {ctr(p,t)}  {point_to_str(p[1])} "
              (x,y),_ = ctr(pts[-1],t,sep=',',raw=True)
              d += f"Q {point_to_str((x,y),sep=',')}, {point_to_str(p[2],sep=',')} "
              d += f"L {point_to_str(b01)} "
              d += f"A {r0},{r0},0,0,0, {point_to_str(b00)}"
              return d
    def GetRingsArcsShardsPositions(\
                              self,
                              rings,
                              arcs,
                              max_graph_sizes,
                              data_scales=[]):
        ring_coords = []
        for i,ring in enumerate(rings):
            r0,r1                     = self.circles_rings_radius_list[ring]
            r1_scaled                 = r0 + max_graph_sizes*(r1-r0)
            rescale_r1                = lambda prct,r1=r1_scaled,r0=r0 : r0+(r1-r0)*prct
            ring_arcs                 = arcs[i]
            ring_n_arcs               = ring_arcs[0]
            ring_n_arcs_shards        = ring_arcs[1]
            ring_arcs_shards_angles   = self.get_drawable_angles(ring_n_arcs_shards)
            d_angle                   = ring_arcs_shards_angles[1]-ring_arcs_shards_angles[0]
            arcs_coords = []
            for arc_id,arc in  enumerate(\
                        ring_arcs[2]):
                shards_coords = []
                for (shard_key,
                     shard_index,
                     shard_label)\
                             in arc :
                    teta            = ring_arcs_shards_angles[shard_index]
                    teta_radian     = math.radians(teta)
                    dteta_radian    = math.radians(teta+d_angle)
                    pdown           = self.get_circle_coords(teta_radian,r0)
                    pup             = self.get_circle_coords(teta_radian,r1_scaled if len(data_scales)==0 else rescale_r1(data_scales[i].get(shard_key,0)) )
                    dpdown          = self.get_circle_coords(dteta_radian,r0)
                    dpup            = self.get_circle_coords(dteta_radian,r1_scaled if len(data_scales)==0 else rescale_r1(data_scales[i].get(shard_key,0)) )
                    start,end       = {'down':pdown,'up':pup},{'down':dpdown,'up':dpup}
                    shard           = {'id':(shard_key,shard_index,shard_label),
                                        'teta':teta,
                                       'teta_radian':teta_radian,
                                       'teta_d_radian':dteta_radian,
                                       'coords':{'start':start,'end':end},
                                       'r1_scaled':r1_scaled if len(data_scales)==0 else rescale_r1(data_scales[i].get(shard_key,0))}
                    shards_coords.append(shard)
                arcs_coords.append({'id':arc_id,
                    'shards':shards_coords})
            ring_coords.append(
                { 'id':ring,
                  'r0':r0,
                  'r1':r1,
                  'r1_scaled':r1_scaled,
                  'arcs':arcs_coords}  )
        self.ring_coords = ring_coords
        return ring_coords
    def ArcSettings(self,
                    ring,
                    arc,
                    data={},
                    thresholds=[],
                    all_key='start',
                    _all=False,
                    with_values=False):
        shard0,shard1 = arc['shards'][0],arc['shards'][-1]
        p0down,p0up   = shard0['coords']['start']['down'],shard0['coords']['start']['up']
        p1down,p1up   = shard1['coords']['end']['down'],shard1['coords']['end']['up']
        all_coords    = [shard['coords'][all_key]['up'] for shard in arc['shards']]
        pts           = ((p0down,p0up,p1up,p1down),all_coords)[_all]
        g             = Group(id=f"ring_{ring['id']}_arc_{arc['id']}_start_{shard0['id'][0]}_end_{shard1['id'][0]}")
        values        = [data.get(shard['id'][0],0) for shard in arc['shards']]
        value         = sum(values)
        range_id = 0
        b0  = 0
        if len(thresholds) >0:
          for _id,b in enumerate(thresholds):
            if value > b0 :
              b0 = b
              if value <= b:
                 range_id = _id
          if value > b : range_id = _id
        range_id_or_values = (range_id,values)[with_values]
        return g,\
              pts,\
              range_id_or_values
    def get_range(self,value,thresholds):
        range_id = 0
        b0  = 0
        if len(thresholds) >0:
          for _id,b in enumerate(thresholds):
            if value > b0 :
              b0 = b
              if value <= b:
                 range_id = _id
          if value > b : range_id = _id
          return range_id
    def add_surface_labels(self,
                          surface_labels,
                          ring,
                          square_angle_left_move,
                          title_font_size,
                          square_size,surface_labels_font_size,
                          surface_labels_font_weight,
                          title_font_weight,fills,
                          strokes,title,
                           surface_labels_plus_n=4):
            n  = len(surface_labels)
            rs = numpy.linspace(ring['r0'],ring['r1'],n+surface_labels_plus_n)
            title_position  = self.get_circle_coords(math.radians(self.label_surface_position-square_angle_left_move),rs[-2])
            Title           = Text(title,insert=title_position,font_size=title_font_size,font_weight=title_font_weight)
            self.image.add(
                Title
            )
            for cpt,r in enumerate(rs[int(surface_labels_plus_n/2):-int(surface_labels_plus_n/2)]):
              label     = surface_labels[cpt]
              f,s      = fills[cpt],strokes[cpt]
              squarep   = self.get_circle_coords(math.radians(self.label_surface_position-square_angle_left_move),r)
              labelp    = (squarep[0]+square_size[0]*2,squarep[1]+square_size[1]#+square_size[1]/2
                           )
              square    = Rect(insert=squarep, size=square_size,stroke=s,fill=f)
              label     = Text(label, insert=labelp,font_size=surface_labels_font_size,font_weight=surface_labels_font_weight)
              self.image.add(\
                square)
              self.image.add(\
                label)
    def ShardSettings(self,
                      ring,
                      arc,
                      shard,
                      data,
                      thresholds,
                      r1_scaled=False,
                      teta=False,
                      other_group_attrs={}):
        shard_key           = shard['id'][0]
        shard_id            = shard_key if len(shard_key)==0 else '_'.join(str(x) for x in shard_key)
        pts                 = (shard['coords']['start']['down'],
                               shard['coords']['start']['up'],
                               shard['coords']['end']['up'],
                               shard['coords']['end']['down'])
        g                   = Group(id=f"ring_{ring['id']}_arc_{arc['id']}_shard_{shard_id}",**other_group_attrs)
        value               = data.get(shard_key,0)
        range_id    = self.get_range(value,thresholds)
        range_or_r1 = (range_id,shard['r1_scaled'])[r1_scaled]
        teta_or_value = (value,(shard['teta_radian'],shard['teta_d_radian']))[teta]
        return g,pts,teta_or_value,range_or_r1
    def AddCirclesDelimiters(self,
                             rings     : list,
                             arcs      : list,
                          linesstyle   : list,
                          labelstyle   : dict  ={},
                      max_graph_sizes  : float =1.,
                      add_labels       : bool  = True,
                      label_left       : float = 40):
                rings_coords           = self.GetRingsArcsShardsPositions(*[rings,arcs,max_graph_sizes])
                for ring in rings_coords:
                    arcs               = ring['arcs']
                    for arc in arcs :
                        shardbegin     = arc['shards'][0]
                        shardend       = arc['shards'][-1]
                        linebegin      = self.drawboundaryline(shardbegin,linesstyle,'start')
                        self.image.add(\
                                linebegin)
                        if add_labels :
                          if self.n_circles-2 == ring['id']:
                            label       = self.get_label(shardbegin,labelstyle,label_left)
                            self.image.\
                              add(label)
                    lineend         = self.drawboundaryline(shardend,linesstyle,'end')
                    self.image.add(\
                        lineend)
                    if add_labels :
                        if self.n_circles-2 == ring['id']:
                            label       = self.get_label(shardend,labelstyle,label_left)
                            self.image.\
                              add(label)

    def AddPieChart(self,
                    ring,
                    arc,
                    data,
                    data_scale_fun = lambda data : dict(map(lambda k : (k,1),data)) ,
                    center_labels=[],
                    fills=[],
                    strokes=[],
                    thresholds=[],
                    title='',
                    title_left_move=10,
                    label_font_size=10,
                    label_font_weight='normal',
                    square_left_move=3,
                    square_size=(3,3),
                    max_graph_sizes = .97,
                    extra_radius=6):
        data_scales         = data_scale_fun(data)
        ring                = self.GetRingsArcsShardsPositions(*[[ring],[arc],max_graph_sizes,[data_scales]])[0]
        for arc in ring\
                   ['arcs'] :
            g,pts,range_id  = self.ArcSettings(ring,arc,data,thresholds)
            f,s             = fills[range_id],strokes[range_id]
            piepath         = Path(d=self.pie(pts,ring['r0'],ring['r1']),fill=f,stroke=s)
            g.add(piepath)
            self.image.add(\
              g)
        if len(center_labels\
                         )>0:
            n               = len(center_labels)
            cx,cy           = self.circles_center
            r0              = ring['r0']
            ys              = numpy.linspace(cy+r0,cy-r0,n+extra_radius)
            title_position  = (cx-title_left_move,ys[-3])
            Title           = Text(title,insert=title_position,font_size=label_font_size,font_weight=label_font_weight)
            self.image.add(
                Title
            )
            for cpt,y in enumerate(ys[3:-3]):
              label         = center_labels[cpt]
              f,s           = fills[cpt],strokes[cpt]
              square_position= (cx-square_left_move,y)
              label_position= self.translate(P=square_size,C=square_position)
              square        = Rect(insert=square_position,size=square_size,fill=f,stroke=s)
              label         = Text(label,insert=label_position,font_size=label_font_size,font_weight=label_font_weight)
              self.image.add(\
                square)
              self.image.add(\
                label)
    def AddCalendar(self,
                    ring,
                    arc,
                    fills,
                    strokes,
                    data,
                    data_scale_fun = lambda data : dict(map(lambda k : (k,1),data)) ,
                    thresholds=[],
                    levels = [],
                    surface_labels=[],
                    surface_labels_font_size=5,
                    surface_labels_font_weight="normal",
                    title='',
                    title_left_move=10,
                    title_font_size=7,
                    title_font_weight='normal',
                    label_font_size=10,
                    label_font_weight='normal',
                    square_angle_left_move=3,
                    square_size=(3,3),
                    max_graph_sizes = .97,
                    extra_radius=6,
                    surface_labels_plus_n=4):
            data_scales=data_scale_fun(data)
            ring       = self.GetRingsArcsShardsPositions(*[[ring],[arc],max_graph_sizes,[data_scales]])[0]
            def get_pts(dr):
                p10,p11 = self.get_circle_coords(teta[0],ring['r0']+dr),self.get_circle_coords(teta[1],ring['r0']+dr)
                return p10,p11
            for arc in ring['arcs']:
              for shard in arc['shards']:
                  g,(p00,_,_,p01),teta,r1_scaled  = self.ShardSettings(ring,arc,shard,data,thresholds,r1_scaled=True,teta=True)
                  unit = (ring['r1'] -ring['r0'])/len(levels)
                  dr = unit
                  for level in levels:
                      value     = data.get((shard['id'][0],level),0)
                      i         = self.get_range(value,thresholds)
                      f,s       = fills[i],strokes[i]
                      if i == 0 :
                        p10,p11 = get_pts(dr)
                        pts     = (p00,p10,p11,p01)
                        bar     = Path(d=self.RectOnArc(pts,ring['r0']),fill=f,stroke=s)
                        dr+=unit
                        p00,p01 = p10,p11
                      else      :
                        p10,p11 = get_pts(dr)
                        pts     = (p00,p10,p11,p01)
                        bar     = Path(d=self.CustomRect(pts),fill=f,stroke=s)
                        dr+=unit
                        p00,p01 = p10,p11
                      g.add(bar)
                      self.image.add(g)
            if len(surface_labels)>0:
                  self.add_surface_labels(
                          surface_labels,
                          ring,
                          square_angle_left_move,
                          title_font_size,
                          square_size,surface_labels_font_size,
                          surface_labels_font_weight,
                          title_font_weight,fills,
                          strokes,title,
                          surface_labels_plus_n)
    def AddBarPlot(self,
                 ring,
                 arc,
                 fills,
                 strokes,
                 data,
                 data_scale_fun = lambda data : dict(map(lambda k : (k,data[k]/max(data.values())),data)) ,
                 thresholds=[],
                 levels = [],
                 surface_labels=[],
                 surface_labels_font_size=5,
                 surface_labels_font_weight="normal",
                 title='',
                 title_left_move=10,
                 title_font_size=7,
                 title_font_weight='normal',
                 label_font_size=10,
                 label_font_weight='normal',
                 square_angle_left_move=3,
                 square_size=(3,3),
                 max_graph_sizes = .97,
                 extra_radius=6,
                 stacked=False,
                 surface_labels_plus_n=4):
        data_scales=data_scale_fun(data)
        ring       = self.GetRingsArcsShardsPositions(*[[ring],[arc],max_graph_sizes,[data_scales]])[0]
        if stacked :
           for arc in ring['arcs']:
              def add_level_to_key(shard,level):
                n =shard.copy()
                k,a,b=n['id']
                n['id']=((k,level),a,b)
                return n
              for shard in arc['shards']:
                  shard          = [self.ShardSettings(ring,arc,add_level_to_key(shard,level),data,thresholds,r1_scaled=True,teta=True) for level in levels ]
                  g               = shard[0][0]
                  for i,(_,\
                        pts,\
                        teta,\
                        r1_scaled) in enumerate(shard):
                      f,s              = fills[i],strokes[i]
                      dr = 0
                      if i == 0 :
                        bar     = Path(d=self.RectOnArc(pts,ring['r0']),fill=f,stroke=s)
                        dr      += (r1_scaled - ring['r0'])
                        prpts   = pts
                      else      :
                        dr      += (r1_scaled - ring['r0'])
                        p10,p11 = self.get_circle_coords(teta[0],ring['r0']+dr),self.get_circle_coords(teta[1],ring['r0']+dr)
                        p00,p01 = prpts[1],prpts[2]
                        pts     = (p00,p10,p11,p01)
                        bar     = Path(d=self.CustomRect(pts),fill=f,stroke=s)
                        prpts   = pts
                      g.add(bar)
                  self.image.add(g)
        else:
          for arc in ring['arcs']:
              for shard in arc['shards']:
                  g,pts,value,range_id    = self.ShardSettings(ring,arc,shard,data,thresholds)
                  f,s                     = fills[range_id],strokes[range_id]
                  bar                     = Path(d=self.RectOnArc(pts,ring['r0']),fill=(f,'none')[value==0],stroke=(s,'none')[value==0])
                  g.add(bar)
                  self.image.add(g)
        if len(surface_labels)>0:
            self.add_surface_labels(
                          surface_labels,
                          ring,
                          square_angle_left_move,
                          title_font_size,
                          square_size,surface_labels_font_size,
                          surface_labels_font_weight,
                          title_font_weight,fills,
                          strokes,title,
                          surface_labels_plus_n)
    def AddBubblePlot(self,
                    ring,
                    arc,
                    fills,
                    strokes,
                    data,
                    data_scale_fun = lambda data : dict(map(lambda k : (k,data[k]/max(data.values())),data)) ,
                    thresholds=[],
                    levels = [],
                    surface_labels=[],
                    surface_labels_font_size=5,
                    surface_labels_font_weight="normal",
                    surface_label_up_move=0,
                    surface_circle_left_move=0,
                    surface_label_right_move=0,
                    title='',
                    title_left_move=10,
                    title_font_size=7,
                    title_font_weight='normal',
                    label_font_size=10,
                    label_font_weight='normal',
                    square_angle_left_move=3,
                    square_size=(3,3),
                    max_graph_sizes = .97,
                    extra_radius=6,
                    surface_labels_plus_n=4,
                    circle_radius=1.5,
                    minsize=1.5):
            data_scales=data_scale_fun(data)
            ring       = self.GetRingsArcsShardsPositions(*[[ring],[arc],max_graph_sizes,[data_scales]])[0]
            positions   = numpy.linspace(ring['r0'],ring['r1'],len(levels)+surface_labels_plus_n)[int(surface_labels_plus_n/2):-int(surface_labels_plus_n/2)]
            for arc in ring['arcs']:
              for shard in arc['shards']:
                  g,(_,
                        (x0,y0),
                        (x1,y1),
                        _),teta,r1_scaled  = self.ShardSettings(ring,arc,shard,data,thresholds,r1_scaled=True,teta=True)
                  shard_width = ((x1-x0)**2+(y1-y0)**2)**0.5
                  for id_,level in enumerate(levels):
                      value       = data.get((shard['id'][0],level),0)
                      prct        = value/max(data.values())
                      i           = self.get_range(value,thresholds)
                      f,s         = fills[id_],strokes[id_]
                      cO          = teta[0] + (teta[1] -teta[0])/2
                      f,s         = (fills[id_],'none')[prct==0],(strokes[id_],'none')[prct==0]
                      p           = positions[id_]
                      r           = minsize
                      r           += shard_width*prct/2
                      cxy         = self.get_circle_coords(cO,p)
                      circle      = Circle(center = cxy, r = r,fill=f,stroke=s)
                      g.add(circle)
                      self.image.add(\
                            circle)
            if len(surface_labels)>0:
              self.add_surface_labels(
                          surface_labels,
                          ring,
                          square_angle_left_move,
                          title_font_size,
                          square_size,surface_labels_font_size,
                          surface_labels_font_weight,
                          title_font_weight,
                          fills,
                          strokes,
                          title,
                          surface_labels_plus_n)
    def AddCurves(self,
                  ring,
                  arc,
                  stroke,
                  fill,
                  data,
                  thresholds=[],
                  data_scale_fun = lambda data : dict(map(lambda k : (k,data[k]/max(data.values())),data)) ,
                  ticks=True,
                  ticks_font_size=5,
                  ticks_font_weight='normal',
                  ticks_up_move=3,
                  ticks_skip=1,
                  ticks_rot=0,
                  t=1/3,
                  fill_opacity=0.7,
                  max_graph_sizes=.97):
          data_scales = data_scale_fun(data)
          ring        = self.GetRingsArcsShardsPositions(*[[ring],[arc],max_graph_sizes,[data_scales]])[0]
          for arc,arc1 in zip(ring['arcs'][:-1],\
                              ring['arcs'][1:]):
              g,pts,values    = self.ArcSettings(ring,arc, all_key='start',_all=True,with_values=True)
              _,boundaries,_   = self.ArcSettings(ring,arc)
              _,(_,
                 pr10,
                 pr11,
                 _),\
                 _  = self.ArcSettings(ring,arc1)
              pts_    = pts +[pr10]
              pts     = [(p0,p,p1) for (p0,p,p1) in zip(pts_[0:-2],pts_[1:-1],pts_[2:])]
              d       = self.curve(pts,boundaries,ring['r0'],t=t)
              curve   = Path(d=d,fill=fill,stroke=stroke)
              g.add(curve)
              self.image.add(g)
              if ticks:
                skip = 0
                for shard in arc['shards']:
                      _,(_,
                         (xi,yi),
                         _,
                         _),(O,_),_  = self.ShardSettings(ring,arc,shard,data,thresholds,r1_scaled=True,teta=True)
                      g_,_,value,_  = self.ShardSettings(ring,arc,shard,data,thresholds,r1_scaled=True,teta=False,other_group_attrs={'transform':f'translate({xi},{yi-ticks_up_move*(-1 if math.degrees(O)<180 else 1)})'})
                      tick =Text(self.human_format(value),
                            text_anchor="middle",
                            transform=f'rotate({math.degrees(O)+ticks_rot})',
                            font_size=ticks_font_size,
                            fill=fill,
                            stroke='none',
                            fill_opacity=fill_opacity,
                            font_weight=ticks_font_weight)
                      if any([skip == ticks_skip,skip==0]):
                        g_.add(tick)
                        skip = 1
                        self.image.add(g_)
          g,pts,_   = self.ArcSettings(ring,arc1, all_key='start',_all=True)
          _,boundaries,_   = self.ArcSettings(ring,arc1)
          pts     = [(p0,p,p1) for (p0,p,p1) in zip(pts[0:-2],pts[1:-1],pts[2:])]
          d       = self.curve(pts,boundaries,ring['r0'],t=t)
          curve   = Path(d=d,fill=fill,stroke=stroke)
          g.add(curve)
          self.image.add(g)

def week_spliter(s,e,group,add_last=False,relative=True):
  s,e = datetime.datetime.fromisoformat(s),datetime.datetime.fromisoformat(e)
  indexes = []
  iso = lambda  d : d.isoformat()[:10]
  label = lambda  d : d.strftime("%d %b")
  is_ = s
  idx = 0
  cpt = 0
  while is_<e :
    ie = min(is_ +  relativedelta(**group) if relative else  datetime.timedelta(**group),e)
    i_idx = []
    dnew = (ie-is_).days
    for day in (range(dnew),range(dnew+1))[add_last]:
      i_idx.append((iso(is_ + datetime.timedelta(days=day)),idx,label(is_ + datetime.timedelta(days=day))))
      idx += 1
    indexes.append(i_idx)
    is_ = ie
    cpt += 1
  if not add_last:
    indexes[-1].append((iso(e),idx,label(e)))
  return (cpt,idx,indexes)
```
